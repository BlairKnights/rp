```python
df['AI Impact'] = df['AI Impact'].str.rstrip('%').astype(float)
```

```python
df['Model/Task Ratio']=df['AI models']/df['Tasks']
```

```python
high_impact=df.loc[df['AI Impact']>=70]
high_impact.head()
```

```python
avg_ratio=df['Model/Task Ratio'].mean()
print(avg_ratio)
df['deviation']=avg_ratio-df['Model/Task Ratio']
high_impact.head()
```
